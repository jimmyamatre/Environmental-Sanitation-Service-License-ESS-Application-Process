**Q15) Identify departments with the highest attrition rate(Resignation %)**

```python
df.head(1)
```

```python
df_resigned_count = df[df['Status'] == 'Resigned'].groupby('Department')['Status'].count()
df_resigned_count
```

```python
df_total_count = df.groupby('Department')['Employee_ID'].count()
attrition_rate = (df_resigned_count / df_total_count) * 100
print("Attrition Rate per Department:")
print(attrition_rate)
```

```python
attrition_rate.sort_values()
```

```python
plt.figure(figsize=(8, 6))
attrition_rate.plot(kind='bar')
plt.title('Attrition Rate per Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation ='vertical', ha ='right')
plt.tight_layout()
plt.show()
```
